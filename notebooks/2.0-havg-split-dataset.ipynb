```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
csv = pd.read_csv("../../data/raw/Customer-Churn-Records.csv")
```

```python
X = csv.drop("Exited", axis=1)
y = csv["Exited"]
X_temp, X_test, y_temp, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=43)
X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, stratify=y_temp, test_size=0.2, random_state=43)
```

```python
train = (X_train, y_train)
test = (X_test, y_test)
val = (X_val, y_val)
```

```python
def save_feature_splits(X_train, X_val, X_test, base_path):  
    X_train.to_csv(f"{base_path}/X_train.csv", index=False)
    X_val.to_csv(f"{base_path}/X_val.csv", index=False)
    X_test.to_csv(f"{base_path}/X_test.csv", index=False)

def save_target_splits(y_train, y_val, y_test, base_path):
    y_train.to_csv(f"{base_path}/y_train.csv", index=False)
    y_val.to_csv(f"{base_path}/y_val.csv", index=False)
    y_test.to_csv(f"{base_path}/y_test.csv", index=False)
```

```python
base_path = '../../data/processed/raw_features'
save_feature_splits(X_train, X_val, X_test, base_path)

base_path = '../../data/processed/target'
save_target_splits(y_train, y_val, y_test, base_path)
```
